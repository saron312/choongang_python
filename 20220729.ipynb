```python
import random
height, weight,score = [],[],[]
for i in range(100):
    height.append(random.randint(100,200))
    weight.append(random.randint(20,100))
    score.append(random.randint(1,1000))
plt.title('키와 몸무게, 체육점수 의 상관관계')
plt.scatter(height, weight, c=height, cmap='rainbow')
plt.colorbar(label='체육점수')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.show()
```

```python
import matplotlib.pyplot as plt
singer = ['A','B','C','D','E']
week1 = [42,58,19,92,84]
week2 = [53,52,48,98,73]
plt.plot(singer, week1)
plt.plot(singer, week2)
plt.show()
```

```python
for i in range(len(week1)):
    week1[i] = -week1[i]
plt.rc('font', family='Malgun Gothic')
plt.title('오디션 프로그램 득표 현황')
plt.barh(singer, week1, label = '첫째 주',color='hotpink')
plt.barh(singer, week2, label= '둘째 주', color='royalblue')
plt.legend()
plt.show()
```
